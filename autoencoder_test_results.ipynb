```python
from fktools import *
```

```python
def draw_labeled_results(points, labels = None):

    plt.scatter(points[:,0], points[:,1],s=0.1)
    plt.title("encoded points")
    plt.show()

    if labels is not None:

        fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(16,9))

        sc00 = ax[0][0].scatter(points[:,0], points[:,1], s=1, c=labels[:,0])
        ax[0][0].set_title("color is number of steps")
        fig.colorbar(sc00, ax=ax[0][0])

        sc01 = ax[0][1].scatter(points[:,0], points[:,1], s=1, c=labels[:,0]%2)
        ax[0][1].set_title("color is number of steps")
        fig.colorbar(sc01, ax=ax[0][1])

        sc10 = ax[1][0].scatter(points[:,0], points[:,1], s=1, c=labels[:,1])
        ax[1][0].set_title("color is start position")
        fig.colorbar(sc10, ax=ax[1][0])

        sc11 = ax[1][1].scatter(points[:,0], points[:,1], s=1, c=labels[:,2])
        ax[1][1].set_title("color is coin phase")
        fig.colorbar(sc11, ax=ax[1][1])
```

```python
points = np.load("autoencoder_8-2-8_encodedpoints.npz")["arr_0"]
labels = np.load("dataset_autoenc_3_y.npz", allow_pickle=True)["arr_0"]

draw_labeled_results(points[labels[:,0]<24], labels[labels[:,0]<24]) # thats a hack, ive run generation first with more than 24 steps
```
